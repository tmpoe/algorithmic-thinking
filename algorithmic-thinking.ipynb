```python
def longest_prefix(a, b):
    for i,c in enumerate(a):
        if i < len(b) and c != b[i]:
            return i
    return -1
        
def longest_postfix(a, b):
    print('post')
    for i in reversed(range(len(b))):
        print(a[i+1], b[i])
        if b[i] != a[i+1]:
            return i+2
    return -1

def solve(a, b):
    
    if len(a) <= len(b) or len(a) > len(b) + 1:
        return [0]
    solution = []

    prefix = a[:longest_prefix(a, b)]
    suffix = a[longest_postfix(a,b):]
    print(prefix, suffix)

    for j in range(len(a)-len(suffix)-2, len(prefix)+1):
        print('j', j, a[:j], a[j+1:], a[:j] + a[j+1:])
        if a[:j] + a[j+1:] == b:
            solution.append(j+1)
    return solution

print(solve('assd', 'asd'))
print(solve('abcdxxxef', 'abcdxxef'))
print(solve('abcdxxxxxef', 'abcdxxxxef'))
        
```

```python
from itertools import count
class Node:
    _ids = count(0)
    def __init__(self, value=0):
        self.value = value
        self.left = None
        self.right = None
        self.id = next(self._ids)
        
    def add_left(self, value):
        if self.left:
            raise Exception("Left already added")
        self.left = Node(value)
        return self.left
        
    def add_right(self, value):
        if self.right:
            raise Exception("Right already added")
        self.right = Node(value)
        return self.right
    
    def __repr__(self):
        
        s = f'{" "*25} Node: {self.id}.'
        s+='\n'
        s+=''.join([' '*(25+i) + '/' + ' '*(5-i)*2 + '\\' + '\n' for i in reversed(range(0, 5))])
        if self.left or self.right:
            s+=f"{' '*20}"
            if self.left:
                s+= f"Id {self.left.id}-v {self.left.value}"
            if self.right:
                if not self.left:
                    s+=' '*8
                s+= f"{' '*5}Id {self.right.id}-v {self.right.value}."
        else:
            s+= f"Leaf node {self.value}"
        s+="\n"*5
        return s
        
            
        
def read_tree(tree_input):
    stack = []
    tree = []
    for i, c in enumerate(tree_input):
        if c == ')':
            root = Node()
            right = stack.pop()
            left = stack.pop()
            if isinstance(left, str) and left.isdigit():
                left = Node(int(left))
            if isinstance(right, str) and right.isdigit():
                right = Node(int(right))

            root.left = left if left != '(' else None
            root.right = right
            if stack:
                _ = stack.pop() # left parenthesis
            stack.append(root)
            tree.append(root)
        else:
            if stack and isinstance(stack[-1], str) and c.isdigit() and stack[-1].isdigit() and tree_input[i-1] != ' ':
                stack[-1] = stack[-1] + c
            elif c != ' ':
                stack.append(c)
    print(stack)
    return tree

def count_all_edges(node):
    if not node.left and node.right:
        return 0
    return (1 + count_all_edges(node.left) if node.left else 0) + (1 + count_all_edges(node.right) if node.right else 0)

def height(node):
    if not node.left and node.right:
        return 0
    
    left = height(node.left) if node.left else 0
    right = height(node.right) if node.right else 0
    
    return max(left, right) + 1

def values(node):
    if node.value:
        return node.value
    
    return (values(node.left) if node.left else 0) + (values(node.right) if node.right else 0)

def print_tree(node):
    
    if not node.left and node.right:
        return
    print(node)
    
    print_tree(node.left) if node.left else None
    print_tree(node.right) if node.right else None

for i in range(5):
    tree_input = input()
    if not tree_input:
        print(0,0)
        continue
    Node._ids = count(0)
    tree = read_tree(tree_input)
    print_tree(tree[-1])
    if not tree:
        print(1, int(tree_input))
        continue
    all_routes_w_returning_home = count_all_edges(tree[-1])*2
    longest_path = height(tree[-1])-1 #-1 for root
    route_wo_returning_home = all_routes_w_returning_home - longest_path
    print(route_wo_returning_home, values(tree[-1]))
'((72 3)(6(((4 9)15)2)))(7 41))'
```
