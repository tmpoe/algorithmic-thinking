```python
# 2.1
from itertools import count
class Node:
    _ids = count(0)
    def __init__(self, value=0):
        self.value = value
        self.left = None
        self.right = None
        self.id = next(self._ids)
        
    def add_left(self, value):
        if self.left:
            raise Exception("Left already added")
        self.left = Node(value)
        return self.left
        
    def add_right(self, value):
        if self.right:
            raise Exception("Right already added")
        self.right = Node(value)
        return self.right
    
    def __repr__(self):
        
        s = f'{" "*25} Node: {self.id}.'
        s+='\n'
        s+=''.join([' '*(25+i) + '/' + ' '*(5-i)*2 + '\\' + '\n' for i in reversed(range(0, 5))])
        if self.left or self.right:
            s+=f"{' '*20}"
            if self.left:
                s+= f"Id {self.left.id}-v {self.left.value}"
            if self.right:
                if not self.left:
                    s+=' '*8
                s+= f"{' '*5}Id {self.right.id}-v {self.right.value}."
        else:
            s+= f"Leaf node {self.value}"
        s+="\n"*5
        return s
        
            
        
def read_tree(tree_input):
    stack = []
    tree = []
    for i, c in enumerate(tree_input):
        if c == ')':
            root = Node()
            right = stack.pop()
            left = stack.pop()
            if isinstance(left, str) and left.isdigit():
                left = Node(int(left))
            if isinstance(right, str) and right.isdigit():
                right = Node(int(right))

            root.left = left if left != '(' else None
            root.right = right
            if stack:
                _ = stack.pop() # left parenthesis
            stack.append(root)
            tree.append(root)
        else:
            if stack and isinstance(stack[-1], str) and c.isdigit() and stack[-1].isdigit() and tree_input[i-1] != ' ':
                stack[-1] = stack[-1] + c
            elif c != ' ':
                stack.append(c)
    print(stack)
    return tree

def count_all_edges(node):
    if not node.left and node.right:
        return 0
    return (1 + count_all_edges(node.left) if node.left else 0) + (1 + count_all_edges(node.right) if node.right else 0)

def height(node):
    if not node.left and node.right:
        return 0
    
    left = height(node.left) if node.left else 0
    right = height(node.right) if node.right else 0
    
    return max(left, right) + 1

def values(node):
    if node.value:
        return node.value
    
    return (values(node.left) if node.left else 0) + (values(node.right) if node.right else 0)

def print_tree(node):
    
    if not node.left and node.right:
        return
    print(node)
    
    print_tree(node.left) if node.left else None
    print_tree(node.right) if node.right else None

for i in range(5):
    tree_input = input()
    if not tree_input:
        print(0,0)
        continue
    Node._ids = count(0)
    tree = read_tree(tree_input)
    print_tree(tree[-1])
    if not tree:
        print(1, int(tree_input))
        continue
    all_routes_w_returning_home = count_all_edges(tree[-1])*2
    longest_path = height(tree[-1])-1 #-1 for root
    route_wo_returning_home = all_routes_w_returning_home - longest_path
    print(route_wo_returning_home, values(tree[-1]))
'((72 3)(6(((4 9)15)2)))(7 41))'
```

```python
# 2.2
from collections import defaultdict
from typing import List
class GenericNode:
    def __init__(self, value=None, children=None):
        self.value: str = value
        self.children: List = children
        self.score: int = 0

    def add_children(self, children: List):
        self.children = children

    def __repr__(self):
        if self.children:
            return f"{self.value}'s children: {' '.join([c.value for c in self.children ])} "
        else:
            return self.value

def get_num_descendants_at_d(children, d, current_depth=1):
    if current_depth == d:
        return len(children)
    return sum([get_num_descendants_at_d(c.children, d, current_depth+1) if c.children else 0 for c in children ])

    
'''
t = int(input())

"Lucas 1 Enzo",
"Zara 1 Amber",
"Sana 2 Gabriel Lucas",
"Enzo 2 Min Becky",
"Kevin 2 Jad Cassie",
"Amber 4 Vlad Sana Ashley Kevin",
"Vlad 1 Omar"
'''
t = 3
nd = [[8, 2], [6,1],[6,2]]
finput = [[
        "Barney 2 Fred Ginger",
        "Ingrid 1 Nolan",
        "Cindy 1 Hal",
        "Jeff 2 Oliva Peter",
        "Don 2 Ingrid Jeff",
        "Fred 1 Kathy",
        "Andrea 4 Barney Cindy Don Eloise",
        "Hal 2 Lionel Mary",
    ],
    [
        "Phillip 5 Jim Phil Jane Joe Paul",
        "Jim 1 Jimmy",
        "Phil 1 Philly",
        "Jane 1 Janey",
        "Joe 1 Joey",
        "Paul 1 Pauly",
    ],
    [
        "Phillip 5 Jim Phil Jane Joe Paul",
        "Jim 1 Jimmy",
        "Phil 1 Philly",
        "Jane 1 Janey",
        "Joe 1 Joey",
        "Paul 1 Pauly",
    ]
]
for i in range(t):
    n, d = nd[i] # list(map(int, input()))
    family_members = dict()
    for j in range(n):
        parent_name, num_children, *children = finput[i][j].split() # input().split()

        children_nodes = []
        for c in children:
            if c not in family_members:
                family_members[c] = GenericNode(value=c)
            children_nodes.append(family_members[c])
        if parent_name not in family_members:
            family_members[parent_name] = GenericNode(value=parent_name, children=children_nodes)
        else:
            family_members[parent_name].add_children(children_nodes)
    final_scores = defaultdict(list)
    for parent_name, parent_node in family_members.items():
        parent_node.score = get_num_descendants_at_d(parent_node.children, d) if parent_node.children else 0
        if parent_node.score > 0:
            final_scores[parent_node.score].append(parent_name)
    
    top_scores = list(sorted(final_scores.keys(), reverse=True))[:3]
    print(f"Tree {i+1}:")
    for score in top_scores:
        final_scores[score] = list(sorted(final_scores[score]))
        for parent in final_scores[score]:
            print(parent, score)
    print()
    


'''
1
7 2
Lucas 1 Enzo
Zara 1 Amber
Sana 2 Gabriel Lucas
Enzo 2 Min Becky
Kevin 2 Jad Cassie
Amber 4 Vlad Sana Ashley Kevin
Vlad 1 Omar
'''
```

```python
# 3.1 Burger fervor
n_min_b = 4
m_min_b = 9
time = 15
memo = {}
def memo_solve(n, m, t):

    if t == 0:
        return 0
    if t-n in memo:
        s1 = memo[t-n]
    else:
        s1 = recursive_solve(n, m, t-n) if t >= n else -1
        memo[t-n] = s1
        
    if t-m in memo:
        s2 = memo[t-m]
    else:  
        s2 = recursive_solve(n, m, t-m) if t >= m else -1
        memo[t-m] = s2

    if s1 == s2 == -1:
        return -1
        
    return max(s1, s2) + 1

result = memo_solve(n_min_b, m_min_b, time)
if result >= 0:
    print(result)
else:
    for i in reversed(range(time)):
        print(i)
        result = memo_solve(n_min_b, m_min_b, i)
        if result >= 0:
            print(result)
            break
```

```python
# 3.2 apples

one_apple_price, m = 1.75, 2
schemes = {3: 4, 2: 2.5}
to_buys = [1, 4]

def recursive_solution(one_apple_price, schemes, to_buy):
    if to_buy <= 0:
        return 0
    else:
        res  = recursive_solution(one_apple_price, schemes, to_buy-1)
    best_val = res + one_apple_price

    for k, v in schemes.items():
        if to_buy - k > -k:
            res = recursive_solution(one_apple_price, schemes, to_buy-k)
            best_val = min(best_val, res + v)

    return best_val

for n in to_buys:
    print(recursive_solution(one_apple_price, schemes, n))
```

```python
def recursive_with_memo_solution(one_apple_price, schemes, to_buy, memo):
    if memo[to_buy] > -1:
        return memo[to_buy]
        
    if to_buy <= 0:
        return 0
    else:
        res  = recursive_with_memo_solution(one_apple_price, schemes, to_buy-1, memo)
    best_val = res + one_apple_price

    for k, v in schemes.items():
        if to_buy - k > -k:
            res = recursive_with_memo_solution(one_apple_price, schemes, to_buy-k, memo)
            best_val = min(best_val, res + v)

    return best_val

for n in to_buys:
    memo = [-1 for _ in range(n+1)]
    print(recursive_with_memo_solution(one_apple_price, schemes, n, memo))
```

```python
#3.3 rivalries

def geese_win_rivalry(geese_results, geese_scores, hawk_results, hawk_scores):
    return geese_results[-1] == 'W' and geese_scores[-1] > hawk_scores[-1]

def geese_loss_rivalry(geese_results, geese_scores, hawk_results, hawk_scores):
    return geese_results[-1] == 'L' and geese_scores[-1] < hawk_scores[-1]

def possible_rivalry_game(geese_results, hawk_results):
    return geese_results[-1] != hawk_results[-1]
    
def recursive_solution(geese_results, geese_scores, hawk_results, hawk_scores, highest=0):
    if len(geese_results) == 0 or len(hawk_results) == 0:
        return 0
    if possible_rivalry_game(geese_results, hawk_results) \
        and geese_win_rivalry(geese_results, geese_scores, hawk_results, hawk_scores) \
        or geese_loss_rivalry(geese_results, geese_scores, hawk_results, hawk_scores):
        highest += geese_scores[-1] + hawk_scores[-1] + recursive_solution(geese_results[:-1], geese_scores[:-1], hawk_results[:-1], hawk_scores[:-1], highest)
    else:
        highest += max(recursive_solution(geese_results, geese_scores, hawk_results[:-1], hawk_scores[:-1], highest), recursive_solution(geese_results[:-1], geese_scores[:-1], hawk_results, hawk_scores, highest))
    return highest

print(recursive_solution('WW', [6, 2], 'LL', [8,1]))
print(recursive_solution('WWW', [6, 2], 'WWW', [8,1]))
print(recursive_solution('WLWW', [3, 4, 1, 8], 'WLLL', [5,1,2,3]))
```
